# Summary
Summary of Interaction
Topic: Distinction between "Standard Error of the Mean" and "Standard Deviation" of Data

Definitions:
Standard Deviation (SD): Measures the dispersion or spread of individual data points around the mean in a dataset. It captures how much individual data points deviate from the average value of the dataset.
Standard Error of the Mean (SEM): Measures the standard deviation of the sample mean’s distribution. It indicates how much the sample mean would vary if multiple samples were drawn from the same population and serves as a measure of how precisely the sample mean estimates the true population mean.
Key Concepts:
Purpose of SD: Describes the variability or spread of data points within the sample.
Purpose of SEM: Reflects the uncertainty or error in using the sample mean as an estimate of the population mean.
Key Differences:
SD is used to describe variability within a single dataset, while SEM is used to describe the variability of the sample mean across different samples.
SD is independent of sample size, whereas SEM is inversely proportional to the square root of the sample size.
This summary captures the detailed explanations and distinctions between SD and SEM that were discussed in the interaction.

# Summary

````python
import numpy as np

# Generate Sample Data and show the 30 samples of data from 10-50
np.random.seed(42) 
sample_data = np.random.normal(loc=50, scale=10, size=30) 

# Compute a 95% bootstrap confidence interval for the mean
def bootstrap_ci(sample, statistic_function=np.mean, n_bootstrap=1000, ci_level=95):
   
    # Generate bootstrap samples and calculate the specified statistic for each sample
    bootstrap_statistics = np.array([statistic_function(np.random.choice(sample, size=len(sample), replace=True))
                                     for _ in range(n_bootstrap)])
    
    # Calculate the percentiles for the confidence interval
    ci_lower = np.percentile(bootstrap_statistics, (100 - ci_level) / 2)
    ci_upper = np.percentile(bootstrap_statistics, 100 - (100 - ci_level) / 2)
    
    return ci_lower, ci_upper

# Calculate and display the 95% bootstrap confidence interval for the mean
mean_ci_lower, mean_ci_upper = bootstrap_ci(sample_data)
print(f"95% Bootstrap Confidence Interval for the Mean: ({mean_ci_lower:.2f}, {mean_ci_upper:.2f})")

# Calculate the confidence interval for the median
# We pass `np.median` as the statistic_function parameter from mean.
median_ci_lower, median_ci_upper = bootstrap_ci(sample_data, statistic_function=np.median)
print(f"95% Bootstrap Confidence Interval for the Median: ({median_ci_lower:.2f}, {median_ci_upper:.2f})")
````

A confidence interval that contains zero indicates that, given the sample data, zero is a likely value for the true population effect. The confidence interval provides information about the range of likely values for the true mean, even in cases when the observed sample mean is not zero. If zero falls inside this range, it suggests that random variation around a true mean of zero may have caused the sample mean to occur. Zero should not be included in the confidence interval in order to reject the null hypothesis. This would imply that there is a meaningful effect that is different from zero and that the genuine effect can only have positive or negative plausible values.
